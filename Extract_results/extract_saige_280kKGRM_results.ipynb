```r
res <- NULL
for(trait in traits){
    message("Busy with trait ", trait)
    inter <- NULL
    for(analysis in analyses){
        file <- paste0('saige_step2_association_', trait, '_chrall_', analysis, '.tsv')
        dat <- fread(file, stringsAsFactors = F, data.table=F)
        dat <- dat[!is.na(dat$max_MAF), ]
        dat <- dat[dat$max_MAF>0.0001, ]
        
        dat$Region <- paste0(dat$Region, ":", trait)
        colnames(dat)[1] <- "Phenotype:Gene"
        #dat$N <- length(which(!is.na(pheno[,trait])))
        #dat$cMAC <- round((dat$AF1) * dat$N *2)
        dat <- dat[dat$MAC>=20, -(c(2:3, 10:12))]
        dat <- dat[!is.na(dat$Pvalue), ]
        dat$LOG10P <- -log10(as.numeric(dat$Pvalue))
        inf_p <- dat[dat$LOG10P==Inf,]
        if(nrow(inf_p)>0){
            for(i in 1:nrow(inf_p)){
                CHISQ <- (inf_p[i, 'BETA_Burden']^2)/(inf_p[i, 'SE_Burden']^2)
                loge_p  <- pchisq(q=CHISQ, df=1, lower.tail=F, log.p=T)
                log10_p <- loge_p * log10(exp(1))
                inf_p[i, 'LOG10P'] <- -1 * log10_p
            }
            dat <- dat[-which(dat$LOG10P==Inf), ]
            dat <- rbind(dat, inf_p)
        }
        colnames(dat)[c(2:(ncol(dat)))] <- paste0(colnames(dat)[c(2:(ncol(dat)))], "__", analysis) 
        if(is.null(inter)){
            inter <- dat
        }else{
            inter <- merge(inter, dat, by="Phenotype:Gene", all=T)
        }
    }
    res <- rbind(res, inter)
}
```

```r
class(res$Pvalue__noPRS) <- class(res$Pvalue__PRSclumped) <- class(res$Pvalue__PRSprscsauto) <- "numeric"
```

```r
#####################################
#### Do some testing; fastGWA
#####################################

#### PRSclumped vs noPRS
rez_table <- NULL
bonferroni <- 7.2e-08
exomewide <- 2.6e-6
comparison <- "noPRS_PRSclumped"
d_list <- NULL
for(ppp in c(bonferroni, exomewide, 1e-5, 1e-4, 1e-3, 0.05)){
  dat <- res[which(res$Pvalue__noPRS<ppp | res$Pvalue__PRSclumped<ppp),]
  dat <- dat[which(!is.na(dat$Pvalue__noPRS)| !is.na(dat$Pvalue__PRSclumped)), ]
  N <- nrow(dat)
  
  ########### test significance difference by paired tests

  # compute the difference
  dat$noPRS <- dat$LOG10P__noPRS
  dat$PRSclumped <- dat$LOG10P__PRSclumped
  dat$d <- dat$PRSclumped - dat$noPRS
  
  d_line <- dat
  d_line$comparison <- comparison
  d_line$varz <- "LOF+missense"
  d_line$cutoff <- ppp
  d_line <- d_line[,c("d", "comparison", "varz", "cutoff")]
  d_list <- rbind(d_list, d_line)
  #shapiro.test(dat$d)
  #shapiro.test(dat[-which(abs(dat$d)>10),'d'])
  #hist(dat$d, breaks = 100) 
  
  ### Wilcoxon paired test
  wilcox.p <- wilcox.test(dat$PRSclumped, dat$noPRS, paired = TRUE)$p.value
  
  ### t-test, remove outliers at 4SD
  sd <- sd(dat$d)
  lower <- mean(dat$d)-(4*sd)
  upper <- mean(dat$d)+(4*sd)
  t.t <- t.test(dat[dat$d > lower & dat$d < upper,'PRSclumped'], dat[dat$d > lower & dat$d < upper,'noPRS'], paired = TRUE)
  t.p <- t.t$p.value
  t.stat <- paste0(round(t.t$estimate,2), " [", round(t.t$conf.int[1],2), "; ", round(t.t$conf.int[2],2), "]") 
  
  line <- c(comparison, ppp, N, wilcox.p, t.p, t.stat)
  
  rez_table <- rbind(rez_table, line)
}
colnames(rez_table) <- c("Comparison", "Significance_Cutoff", "N", "Wilcoxon.P", "Ttest.P", "Ttest.Estimate")
rez_table_tot <- as.data.frame(rez_table)
rez_table_tot$Model <- "SAIGE_GENEplus"
```
